```python
from sklearn.model_selection import train_test_split

x_lr = df1.drop(columns=['y','euribor3m','nr.employed'],axis = 1)
y_lr = df1.y

x_train_lr,x_test_lr,y_train,y_test = train_test_split(x_lr,y_lr,test_size=0.20,random_state=42)
```

```python
x_train_lr.shape
```

```python
x_test_lr.shape
```

```python
cat_col = ['job','marital','default','housing','loan','contact','day_of_week','month']
minmax_col = ['emp.var.rate','cons.price.idx','cons.conf.idx']
stand_col = ['pdays','previous_log']
robust_col = ['age','campaign_log']
ordinal_col = ['education','poutcome']

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler,StandardScaler,RobustScaler,OneHotEncoder,OrdinalEncoder

Transformer = ColumnTransformer([
    ('cat_encode',OneHotEncoder(drop = 'first'),cat_col),
    ('minmax_scalar',MinMaxScaler(),minmax_col),
    ('stand_scalar',StandardScaler(),stand_col),
    ('robust_scalar',RobustScaler(),robust_col),
    ('ordinal_encode',OrdinalEncoder(),ordinal_col)
])

x_train_lr_trans = Transformer.fit_transform(x_train_lr)
x_test_lr_trans = Transformer.transform(x_test_lr)
```

```python
x_train_lr_trans.shape
```

```python
x_test_lr_trans.shape
```

### Convert y column to numericals

```python
dict = {
    'yes' : 1,
    'no' : 0
}

y_train = y_train.map(dict)
y_test = y_test.map(dict)
```

```python
y_train.shape
```

```python
y_test.shape
```

# Task 2:- Predictive model which will help the bank marketing team.

## Logistic Regession Model Building

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,roc_curve,confusion_matrix,ConfusionMatrixDisplay

lr = LogisticRegression(random_state=10)

lr.fit(x_train_lr_sm,y_train_lr_sm)
```

#### Prediction

```python
y_pred_lr = lr.predict(x_test_lr_trans)
```

### Evolution

```python
from sklearn.metrics import roc_auc_score, roc_curve

y_proba_lr = lr.predict_proba(x_test_lr_trans)[:,1]

fpr, tpr, thresholds = roc_curve(y_test, y_proba_lr)

plt.plot(fpr, tpr, label=f"AUC={roc_auc_score(y_test,y_proba_lr):.3f}")
plt.plot([0,1],[0,1],'k--')
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC Curve - Logistic Regression")
plt.legend()
plt.show()
```

```python
print(classification_report(y_test,y_pred_lr))

cm = confusion_matrix(y_test, y_pred_lr)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.title("Confusion Matrix")
plt.show()

#Roc AUc
```

```python
cat_col = ['job','marital','default','housing','loan','contact','day_of_week','month']
minmax_col = ['emp.var.rate','cons.price.idx','cons.conf.idx','euribor3m','nr.employed']
stand_col = ['pdays','previous_log']
robust_col = ['age','campaign_log']
ordinal_col = ['education','poutcome']

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler,StandardScaler,RobustScaler,OneHotEncoder,OrdinalEncoder

Transformer = ColumnTransformer([
    ('cat_encode',OneHotEncoder(drop = 'first'),cat_col),
    ('minmax_scalar',MinMaxScaler(),minmax_col),
    ('stand_scalar',StandardScaler(),stand_col),
    ('robust_scalar',RobustScaler(),robust_col),
    ('ordinal_encode',OrdinalEncoder(),ordinal_col)
])

x_train_trans = Transformer.fit_transform(x_train)
x_test_trans = Transformer.transform(x_test)
```
